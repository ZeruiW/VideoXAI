```python
import os
import shutil
import random
from collections import defaultdict

# 
label_file = 'kinetics400_val_list_videos.txt'
source_dir = 'videos_val'
target_dir = 'videos_test'

# 
n_videos_per_label = 3  # 

# 
os.makedirs(target_dir, exist_ok=True)

# 
label_dict = defaultdict(list)
with open(label_file, 'r') as f:
    for line in f:
        video_name, label = line.strip().split()
        label = int(label)
        label_dict[label].append(video_name)

# n
selected_videos = {}
for label in range(400):  # 0399
    if label in label_dict:
        videos = label_dict[label]
        n_select = min(n_videos_per_label, len(videos))  # 
        selected_videos[label] = random.sample(videos, n_select)

# 
copied_count = 0
for label, video_names in selected_videos.items():
    for video_name in video_names:
        source_path = os.path.join(source_dir, video_name)
        target_path = os.path.join(target_dir, video_name)
        
        if os.path.exists(source_path):
            shutil.copy2(source_path, target_path)
            print(f"Copied {video_name} (Label: {label}) to {target_dir}")
            copied_count += 1
        else:
            print(f"Warning: {video_name} not found in {source_dir}")

# 
print(f"Total videos copied: {copied_count}")
print(f"Number of labels represented: {len(selected_videos)}")
```
